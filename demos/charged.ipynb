# Geodesic Demo

First, we import the necessary modules.

```python
%load_ext autoreload
%autoreload 2

from fadge import GRRT
from jax   import numpy as np

from matplotlib import pyplot as plt
from itertools  import product
from math       import sqrt
```

```python
fs = {}

for a, q in [
    (-0.9,0),
    (-0.8,0.5),
    (-0.7,0.7),
    (0,0),
    (0,0.5),
    (0,1),
    (0.9,0),
    (0.8,0.5),
    (0.7,0.7),
]:
    grrt = GRRT(aspin=a, qcharge=q, atol=1e-3, rtol=0, names={'ind':'lambda'})
    grrt.set_cam(1e4, 90, 0)
    grrt.set_axis(fov=16, n=32)
    l, f = grrt.geode()    
    fs[(a, q)] = f
```

```python
def D(q): # assume spin is zero
    x = sqrt(9 - 8 * q * q) + 3
    d = x * x / sqrt(2 * (x - 2 * q * q))
    return [0.5*d, -0.5*d]
```

```python
fig, axes = plt.subplots(3, 3, figsize=(18,18))

[axes[1,0].axhline(y) for y in D(0)]
[axes[1,1].axhline(y) for y in D(0.5)]
[axes[1,2].axhline(y) for y in D(1)]

for i, ((a,q),f) in enumerate(fs.items()):
    fx = f[:,0,1]
    fy = f[:,0,2]
    fz = f[:,0,3]
    m  = fx[0,:]*fx[0,:] + fy[0,:]*fy[0,:] + fz[0,:]*fz[0,:] > 10

    ax = axes[divmod(i,3)]
    for k in range(32):
        ax.plot(fx[:,k], fy[:,k], '-', color='r' if m[k] else 'k')

    ax.set_title(f'a={a}, q={q}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_xlim(-20,20)
    ax.set_ylim(-20,20)
    ax.set_aspect('equal')
```
